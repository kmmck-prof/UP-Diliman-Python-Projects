# EEE 121 Week 1 - Pig Latin

Pig Latin is a language game in which English words are altered, usually by manipulating the position of the letters in words and/or adding suffixes/prefixes. Here are the rules so that you won't get lost in translation.

1. If a word starts with a consonant and a vowel, put the first letter of the word at the end of the word and add "ay."
2. If a word starts with two consonants move the two consonants to the end of the word and add "ay."
3. If a word starts with a vowel add the word "way" at the end of the word.

# Solution

````python
# If a word starts with a consonant and a vowel, put the first letter of the word at the end of the word and add "ay."
# If a word starts with two consonants move the two consonants to the end of the word and add "ay."
# If a word starts with a vowel add the word "way" at the end of the word.

def piglatin_translate(phrase):
    vowels = ['a','e','i','o','u']
    new_words = []
    words = phrase.split(" ")
    for word in words:
        letters = word.lower()
        if letters[0] not in vowels and letters[1] in vowels:
            letters = letters[1:] + letters[0] + "ay"
        elif letters[0] not in vowels and letters[0] not in vowels:
            letters = letters[2:] + letters[:2] + "ay"
        elif letters[0] in vowels:
            letters += "way"
        new_words.append(letters)

    return " ".join(new_words)

if __name__ == "__main__":
    translated_phrases = []
    for x in range(3):
        input_line = str(input())
        translated_phrases.append(piglatin_translate(input_line))
    
    for piglatin_phrase in translated_phrases:
        print(piglatin_phrase)
````
